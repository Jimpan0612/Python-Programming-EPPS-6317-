## Opening a shapefile in Fiona

```python
import fiona
```

```python
rroads = fiona.open("home/Railroads.shp")
rroads
```

```python
# Get first feature
f = rroads.next()
# What geometry type is it?
f['geometry']['type']
```

```python
# Get the shapefile extent
rroads.bounds
```

```python
rroads.schema
```

```python
# Use shapefile properties in conditional logic
if rroads.schema['geometry'] == 'Point':
    print("This is a point file")
elif rroads.schema['geometry'] == 'LineString' or rroads.schema['geometry'] == 'MultiLineString':
    print("This is a line file")
elif rroads.schema['geometry'] == 'Polygon' or rroads.schema['geometry'] == 'MultiPolygon':
    print("This is a polygon file")
else:
    print("I don't know *what* this is!")
```

## Working with shapefile attribute data

```python
# Open Countries shapefile
countries = fiona.open("home/Countries.shp")
# Get first feature
f = next(countries)
# Take a look at first feature's attributes so we can get an idea of the fieldnames
f['properties']
```

```python
# Loop through all countries (features) in the shapefile
for c in countries:
    # Include Country's name and population in printout)
    print(f"{c['properties']['CNTRY_NAME']} has a population of {c['properties']['POP_CNTRY']}.")
```

## Opening Landsat imagery with rasterio

```python
raster = rasterio.open("home/LE70220492002106EDC00_stack.gtif")
raster
```

## Map algebra with rasterio

```python
dem = rasterio.open("home/usdem10k.tiff")
dem_data = dem.read(1)
dem_data
```

```python
import matplotlib.pyplot as plt
```

```python
# Set up a matplotlib figure for DEM display
fig, ax = plt.subplots(figsize=(12,8))
# Set up DEM plot with specified data source and title
demplot = show(dem, ax=ax, title='US Elevation in Feet')
cbar = fig.colorbar(ax.imshow(dem_data), fraction=0.035, pad=0.01)
plt.show()
```

```python
dem_data_m = dem_data * 0.3048
```

```python
# Set up a matplotlib figure for DEM display
fig, ax = plt.subplots(figsize=(12,8))
# Set up DEM plot with specified data source and title
demplot = show(dem, ax=ax, title='US Elevation in Meters')
cbar = fig.colorbar(ax.imshow(dem_data_m), fraction=0.035, pad=0.01)
plt.show()
```
