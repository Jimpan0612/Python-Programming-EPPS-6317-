```python
import re
import pandas as pd

# Function to convert DMS coordinates to decimal degrees
def dms_to_decimal(coord, direction):
    degrees, minutes = map(int, re.findall(r'\d+', coord))
    decimal_degrees = degrees + minutes / 60.0
    return decimal_degrees if direction in ('N', 'E') else -decimal_degrees

# Read worldcities.txt
with open('worldcities.txt') as file:
    data = file.read()

# Define a pattern for extracting DMS coordinates
pattern = re.compile(r"([A-Za-z\s,]+)\s+(\d+°\d+'[NS])\s+(\d+°\d+'[EW])")

# Extract coordinates and convert to decimal degrees
coordinates = []
for match in pattern.finditer(data):
    city, lat, lon = match.groups()
    coordinates.append((city.strip(), dms_to_decimal(lat, lat[~lat.isdigit()]), dms_to_decimal(lon, lon[~lon.isdigit()])))

# Create a DataFrame
df = pd.DataFrame(coordinates, columns=['City', 'Latitude', 'Longitude'])

# Print the DataFrame
print(df)
```
