```python
import pandas as pd

# (In thousands of US dollars)

df_2013 = pd.read_excel('2013_con_quarter04_bs.xlsx')
df_2015 = pd.read_excel('2015_con_quarter04_bs.xlsx')
df_2017 = pd.read_excel('2017_con_quarter04_bs.xlsx')
df_2019 = pd.read_excel('2019_con_quarter04_bs.xlsx')
df_2021 = pd.read_excel('2021_con_quarter04_bs.xlsx')
df_2022 = pd.read_excel('2022_con_quarter04_bs.xlsx')

print(df_2013.head(15))
```

```python
# Choose columns need
df_2013.columns = ['Column1', 'Column2', 'Column3', 'Column4','Column5','Column6']
selected_cols = ['Column1', 'Column5', 'Column6']
df_2013 = df_2013[selected_cols]

df_2013
```

```python
# Rename columns, Select specific columns, drop NA
def process_yearly_data(df):
    # Rename columns
    df.columns = ['Column1', 'Column2', 'Column3', 'Column4', 'Column5', 'Column6']
    
    # Select specific columns
    selected_cols = ['Column1', 'Column5', 'Column6']
    df_processed = df[selected_cols]
    
    # drop NA
    df_processed = df_processed.dropna()
    
    return df_processed
```

```python
df_2013 = df_2013.dropna()
df_2013
```

```python
new_column_names = {'Column1': 'samsung', 'Column5': '2013', 'Column6': '2012'}
df_2013_2012 = df_2013.rename(columns=new_column_names)
df_2013_2012
```

```python
new_column_order = ['samsung', '2012', '2013']
df_2012_2013 = df_2013_2012[new_column_order]
df_2012_2013
```

```python
# change all

df_2015 = process_yearly_data(df_2015)
df_2017 = process_yearly_data(df_2017)
df_2019 = process_yearly_data(df_2019)
df_2021 = process_yearly_data(df_2021)
df_2022 = process_yearly_data(df_2022)

df_2015
```

```python
new_column_names = {'Column1': 'samsung', 'Column5': '2015', 'Column6': '2014'}
df_2015_2014 = df_2015.rename(columns=new_column_names)

new_column_order = ['samsung', '2014', '2015']
df_2014_2015 = df_2015_2014[new_column_order]
df_2014_2015
```

```python
new_column_names = {'Column1': 'samsung', 'Column5': '2017', 'Column6': '2016'}
df_2017_2016 = df_2017.rename(columns=new_column_names)

new_column_order = ['samsung', '2016', '2017']
df_2016_2017 = df_2017_2016[new_column_order]
df_2016_2017
```

```python
new_column_names = {'Column1': 'samsung', 'Column5': '2019', 'Column6': '2018'}
df_2019_2018 = df_2019.rename(columns=new_column_names)

new_column_order = ['samsung', '2018', '2019']
df_2018_2019 = df_2019_2018[new_column_order]
df_2018_2019
```

```python
new_column_names = {'Column1': 'samsung', 'Column5': '2021', 'Column6': '2020'}
df_2021_2020 = df_2021.rename(columns=new_column_names)

new_column_order = ['samsung', '2020', '2021']
df_2020_2021 = df_2021_2020[new_column_order]
df_2020_2021
```

```python
new_column_names = {'Column1': 'samsung', 'Column5': '2022', 'Column6': '2021'}
df_2022_2021 = df_2022.rename(columns=new_column_names)

new_column_order = ['samsung', '2021', '2022']
df_2021_2022 = df_2022_2021[new_column_order]
df_2021_2022
```

```python
selected_cols = ['samsung', '2022']
df_2022 = df_2021_2022[selected_cols]

df_2022
```

```python
selected_row_names = ['Total assets', 'Total liabilities', 'Total liabilities and equity']

selected_rows = []

# Loop through each DataFrame and select rows based on the specified row names
for df in [df_2012_2013, df_2014_2015, df_2016_2017, df_2018_2019, df_2020_2021, df_2022]:
    # Use the 'isin' method to check if the 'samsung' column matches the selected row names
    selected_rows_df = df[df['samsung'].isin(selected_row_names)]
    
    # Append the selected rows DataFrame to the list
    selected_rows.append(selected_rows_df)

# Concatenate the selected rows along the rows axis
df = pd.concat(selected_rows, axis=0, ignore_index=True)
df
```

```python
# Group by the 'samsung' column and sum the values for each group
grouped_df = df.groupby('samsung').sum().reset_index()

# Transpose the DataFrame to switch rows and columns
combined_df = grouped_df.T.reset_index()

# Set the first row as column headers
combined_df.columns = combined_df.iloc[0]

# Drop the first row (which is now duplicated as column headers)
combined_df = combined_df[1:].reset_index(drop=True)

# Rename the index column
combined_df = combined_df.rename(columns={'samsung': 'Year'})

# Convert the 'Year' column to a string for better representation
combined_df['Year'] = combined_df['Year'].astype(str)

# Display the resulting DataFrame
combined_df
```

```python
new_column_names = {'Total assets': 'Samsung Total assets', 'Total liabilities': 'Samsung Total liabilities', 'Total liabilities and equity': 'Samsung Total liabilities and equity'}
combined_df = combined_df.rename(columns=new_column_names)
combined_df
```

```python
combined_df.to_csv('Samsung_FS.csv', index=False)
```
